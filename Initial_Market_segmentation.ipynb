```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by state and calculate average modal price
state_price_summary = df.groupby('state')['modal_price'].mean().sort_values(ascending=False)

# Top 10 states by average modal price
top_states = state_price_summary.head(10)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=top_states.values, y=top_states.index, palette="viridis")
plt.title("Top 10 States by Average Modal Price")
plt.xlabel("Average Modal Price (INR)")
plt.ylabel("State")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Top 10 districts by average modal price
district_price_summary = df.groupby('district')['modal_price'].mean().sort_values(ascending=False)
top_districts = district_price_summary.head(10)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=top_districts.values, y=top_districts.index, palette="mako")
plt.title("Top 10 Districts by Average Modal Price")
plt.xlabel("Average Modal Price (INR)")
plt.ylabel("District")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
